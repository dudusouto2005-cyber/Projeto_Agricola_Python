# Monitoramento da Umidade do Solo

## Informações do experimento

**Cultura:** Tomate (*Solanum lycopersicum*)

**Material genético:** San Marzano

**Área experimental:** 42 m²

**Sistema de cultivo:** Campo aberto

**Irrigação:** Gotejamento

## Objetivo

Avaliar a umidade do solo durante sete dias consecutivos para verificar se o manejo da irrigação está mantendo condições adequadas para o desenvolvimento das plantas.

A umidade do solo é um dos principais fatores relacionados ao crescimento vegetal, pois influencia diretamente a absorção de água e nutrientes pelas raízes.

```python
# ===========================================
# MONITORAMENTO DA UMIDADE DO SOLO
# Cultura: Tomate San Marzano
# Área: 42 m²
# ===========================================

# Umidade (%) observada durante sete dias

umidade = [31, 29, 35, 33, 28, 30, 32]

media = sum(umidade) / len(umidade)
maior = max(umidade)
menor = min(umidade)

print("RELATÓRIO DE UMIDADE DO SOLO")
print("-" * 40)

for dia, valor in enumerate(umidade, start=1):
    print(f"Dia {dia}: {valor}%")

print("-" * 40)
print(f"Média semanal: {media:.1f}%")
print(f"Maior valor: {maior}%")
print(f"Menor valor: {menor}%")

if media >= 30:
    print("\nCondição do solo: ADEQUADA")
else:
    print("\nCondição do solo: NECESSITA IRRIGAÇÃO")
```

# Interpretação

O monitoramento foi realizado em uma área experimental de 42 m² cultivada com tomate da variedade San Marzano.

Durante os sete dias de avaliação, a umidade variou entre 28% e 35%, apresentando média de aproximadamente 31%.

Esses resultados indicam que o solo permaneceu dentro da faixa considerada adequada para o desenvolvimento inicial da cultura.

A manutenção dessa umidade favorece o crescimento radicular, melhora a absorção de nutrientes e reduz o estresse hídrico das plantas.

Esse tipo de análise permite que o produtor decida o momento correto para irrigar, aumentando a eficiência do uso da água e reduzindo desperdícios.
